# Visualization of Appendix A.1

Load packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import matplotlib.dates as md
import matplotlib.dates as mdates
```

Load the data

```python
df = pd.read_csv('../data/2_final_output/daily_mobility_metrics_from_mobile_phone_data.csv', delimiter = ';',decimal=",")
# convert timestamp column to datetime format
df['date'] = pd.to_datetime(df['date'])
# set index
df.set_index('date')
# calculate trend of mobility metrics using a 28 day moving average
df['user_count_mov_avg'] = df['user_count'].rolling(28).mean().shift(-14)
df['weekly_penetration_rate_mov_avg'] = df['weekly citywide penetration rate'].rolling(28).mean().shift(-14)
df['number_movements_mov_avg'] = df['amount of movements'].rolling(28).mean().shift(-14)
df['distance_mov_avg'] = df['average distance'].rolling(28).mean().shift(-14)
df['modularity_mov_avg'] = df['Modularity'].rolling(28).mean().shift(-14)
df['residential_mov_avg'] = df['Residential'].rolling(28).mean().shift(-14)
df['radius_mov_avg'] = df['Radius of Gyration'].rolling(28).mean().shift(-14)
```

Plot the temporal singal of 'unique users' 

```python
fig, ax = plt.subplots(figsize = (10, 5))
plt.rcParams["font.family"] = "Times New Roman"
sns.lineplot(ax = ax, x='date', y='user_count', data=df, color='orange',label ='Unique user count')
sns.lineplot(ax = ax, x='date', y='user_count_mov_avg', data=df, color='darkred', label ='Unique user count [28 day moving average]')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b\n%Y'))
plt.setp(ax.xaxis.get_majorticklabels(), size= 15)
plt.setp(ax.yaxis.get_majorticklabels(), size= 15)
ax.tick_params(axis = 'x', which = 'major', length = 7)
ax.tick_params(axis = 'x', which = 'minor', length = 20)
plt.xlabel('Date', size=18)
plt.ylabel('Unique user count [million]', size=18)
ax.legend(prop={'size': 15})
ax.grid(True, color='white')
ax.set_facecolor('gainsboro')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
plt.rcParams["font.family"] = "Times New Roman"
sns.lineplot(ax = ax, x='date',  y='amount of movements', data=df, color='orange',label ='Amount of movements')
sns.lineplot(ax = ax, x='date', y='number_movements_mov_avg', data=df, color='darkred', label ='Amount of movements [28 day moving average]')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b\n%Y'))
plt.setp(ax.xaxis.get_majorticklabels(), size= 15)
plt.setp(ax.yaxis.get_majorticklabels(), size= 15)
ax.tick_params(axis = 'x', which = 'major', length = 7)
ax.tick_params(axis = 'x', which = 'minor', length = 20)
plt.xlabel('Date', size=18)
plt.ylabel('Amount of movements [million]', size=18)
ax.legend(prop={'size': 15})
ax.grid(True, color='white')
ax.set_facecolor('gainsboro')
plt.show()
```
