```python
import copy
import glob
import skmob
import numpy as np
import pandas as pd
import seaborn as sns
import geopandas as gpd
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.colors as pltCols
from datetime import datetime as dt
import skmob.measures.individual as sk_id
from matplotlib.ticker import PercentFormatter
from mpl_toolkits.axes_grid1 import make_axes_locatable

sns.set(rc={'axes.facecolor':'lightgray', 'figure.facecolor':'white'})
plt.rcParams['font.family'] = 'Times New Roman'
```

# Data Inspection
The cleaned data is inspected here. The Notebook provides an overlook over the data before we calculate the actual measures for different subsets during the loop.

```python
tweets_pois = pd.read_csv(r'data\tweets\preprocessed_tweets_with_poi_location.csv')
tweets_pois['Timestamp'] = tweets_pois['Timestamp'].apply(lambda x: dt.strptime(x, "%Y-%m-%d %H:%M:%S"))
tweets = tweets_pois

barrios = gpd.read_file(r'data\shps\neighborhoods.shp')
```

```python
tdf = skmob.TrajDataFrame(tweets, latitude='lat', longitude='lon', datetime='Timestamp', user_id='User_ID')
jl = sk_id.jump_lengths(tdf).jump_lengths

all_moves = []
jl.apply(lambda x: all_moves.extend(x))
all_moves = pd.Series(all_moves)
```

```python
# exclude all 0 movements
all_moves = all_moves[all_moves > 0]
```

### Displacements distances
How far are the displacements?

```python
bins = np.arange(0, 60, 1) # fixed bin size

fig, ax = plt.subplots(1)
ax.set_title('Number of Displacements by km bracket (excluding 0)')
ax.set_xlabel('Displacement Distance')
ax.set_ylabel('Number of Displacements')
ax.hist(all_moves, bins=bins, weights=np.ones_like(all_moves)*100 / len(all_moves), alpha=0.5, ec='black', color='black')
ax.yaxis.set_major_formatter(PercentFormatter())
```

### USED IN PAPER: Displacements times
How much time is inbetween a displacement? The Graph shows a cumulative plot. The largest interval of displacements that we derive our metrics from is 31 days (where the dashed line is), more detail on that is shown in the subsequent notebooks. This plot is featured in the published paper.

```python
wt = sk_id.waiting_times(tdf).waiting_times
all_times = []
wt.apply(lambda x: all_times.extend(x))
all_times = pd.Series(all_times) / 86400
```

```python
bins = [len(all_times[all_times < i]) / len(all_times) * 100 for i in np.arange(50)]
```

```python
fig, ax = plt.subplots(1)
ax.set_xlabel('Inter-tweet time [days]', size=15)
ax.set_ylabel('Relavive number of displacements', size=15)
ax.plot(range(50), bins, color='black')
ax.set_xticks(np.arange(0, 55, 5))
plt.axvline(x=31, color='black', linestyle='--')
ax.legend(['Cumulative number of displacements', '31 days inter-event time'], fontsize=15)

ax.tick_params(axis='x', which='major', labelsize=15)
ax.tick_params(axis='y', which='major', labelsize=12)

ax.yaxis.set_major_formatter(PercentFormatter())

# figure2_3
```

### Tweet locations
An overview on the locations that have tweets.

```python
gdf = copy.deepcopy(tweets)
gdf['wkt'] = gpd.GeoSeries.from_wkt(gdf.wkt)
gdf = gpd.GeoDataFrame(gdf, geometry='wkt')
```

```python
gdf_no_d = gdf.drop_duplicates(subset='wkt')

fig, ax = plt.subplots(1, figsize=(18, 12))
ax.set_title('All Tweet Locations during the study period')
plt.xticks([])
plt.yticks([])
barrios.boundary.plot(ax=ax, color='black', linewidth=0.5)
gdf_no_d.plot(ax=ax, color='blue', marker='s', markersize=8, edgecolor='black', linewidth=0.5)
```

### Hourly Heatmap GIF
This GIF shows an hourly Kernel Density Esimation of tweets in Rio for the whole timeframe.

```python
xlims = barrios.dissolve().bounds[['minx', 'maxx']].values[0] + [-0.05, 0.05]
ylims = barrios.dissolve().bounds[['miny', 'maxy']].values[0] + [-0.05, 0.05]

gdf['hour'] = gdf.Timestamp.apply(lambda x: x.hour)
grouped = gdf.groupby('hour')

heats = []
for i in range(24):
    vals = grouped.get_group(i)
    fig, ax = plt.subplots(1, figsize=(12, 8))
    ax.set_title('Tweet Heatmap for {}:00 day hours'.format(i))
    plt.xticks([])
    plt.yticks([])
    plt.xlim(xlims)
    plt.ylim(ylims)
    barrios.boundary.plot(ax=ax, color='black', linewidth=0.5)
    sns.kdeplot(data=vals,
                x='lon', 
                y='lat', 
                fill=True,
                cmap='coolwarm',
                alpha=0.8, 
                gridsize=200,
                levels=20,
                ax=ax)
    
    heats.append(fig)
    
    i = str(i)
    if len(i) == 1:
        i = '0' + i

    plt.savefig(f'result_plots/hourly_twitter_heatmap/{i}heatmap.png')
    plt.close()
    print(f'Saved figure {i}')
```

```python
fp_in = "result_plots/hourly_twitter_heatmap/*heatmap.png"
fp_out = "result_plots/hourly_twitter_heatmap/hourly_heatmap.gif"

imgs = (Image.open(f) for f in sorted(glob.glob(fp_in)))
img = next(imgs)  # extract first image from iterator
img.save(fp=fp_out, format='GIF', append_images=imgs,
         save_all=True, duration=200, loop=0)
```

### Tweets per Barrio

```python
gdf.cod = gdf.cod.apply(str)
tweets_per_barr = gdf.groupby('cod').count().drop(columns=['Timestamp'])
tweets_per_barr = pd.DataFrame(tweets_per_barr['Tweet_ID'])

barrios.CODBAIRRO = barrios.CODBAIRRO.astype(int)
barrios = barrios.set_index('CODBAIRRO')

tweets_per_barr.index = tweets_per_barr.index.astype(int)
counts_with_geom = tweets_per_barr.join(barrios).rename(columns={'Tweet_ID': 'count'})[['count', 'popsize', 'geometry', 'NOME']]

counts_with_geom = gpd.GeoDataFrame(counts_with_geom, geometry='geometry')
counts_with_geom['rel_counts_100000'] = counts_with_geom['count'] / counts_with_geom.popsize * 100000
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))
ax.set_title('Tweets per barrio during the study period (absolute)')
plt.xticks([])
plt.yticks([])
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
counts_with_geom.plot(ax=ax, column='count', legend=True, cax=cax)


fig, ax = plt.subplots(1, figsize=(12, 8))
ax.set_title('Tweets per barrio per 100.000 inhabitants during the study period (log)')
plt.xticks([])
plt.yticks([])
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
counts_with_geom.plot(ax=ax, column='rel_counts_100000', norm=pltCols.LogNorm\
                      (vmin=counts_with_geom.rel_counts_100000.min(), vmax=counts_with_geom.rel_counts_100000.max()),\
                      legend=True, cax=cax)
```

### Barrios by Tweet Volume
Barrios are sorted ascending by their relative tweet volume.

```python
tweets_per_barr_perc = copy.deepcopy(tweets_per_barr)
tweets_per_barr_perc.Tweet_ID = tweets_per_barr_perc.Tweet_ID / tweets_per_barr_perc.Tweet_ID.sum() * 100
tweets_per_barr_perc = pd.DataFrame(tweets_per_barr_perc.Tweet_ID)
```

```python
fig, ax = plt.subplots(1)
ax.set_xticks([])
ax.set_xlabel('Neighborhoods sorted by tweet volume')
ax.set_ylabel('Share of tweets in barrio [%]')
ax.set_title('Tweet share by neighborhood (ascending order)')
plt.plot(sorted(tweets_per_barr_perc.Tweet_ID.tolist()), color='black')

fig, ax = plt.subplots(1)
ax.set_xticks([])
ax.set_xlabel('Neighborhoods sorted by tweet volume')
ax.set_ylabel('Share of Tweets in Barrio [%]')
ax.set_title('Tweet share by neighborhood (ascending order) (cumulative)')
plt.plot(np.cumsum(sorted(tweets_per_barr_perc.Tweet_ID.tolist())), color='black')
```
